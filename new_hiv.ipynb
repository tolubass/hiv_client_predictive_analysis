```python
# Dictionary mapping old names to new ones
rename_dict = {
    'Mstatus': 'Marital_Status',
    'Cat_Perc': 'Perception_Category',
    'years': 'Years_on_Treatment',
    'Over_Comf': 'Overall_Comfort_Level',
    'Edulevel': 'Education_Level',
    'monthIncome': 'Monthly_Income',
    'numb': 'Household_Size',
    'treatmentreg': 'Treatment_Regimen',
    'TimRecCare': 'Care_Timeliness',
    'comfortability_copy7': 'Comfort_StaffInteraction'
}

# Apply the renaming
df1.rename(columns=rename_dict, inplace=True)

# Optional: check the changes
print(df1[list(rename_dict.values())].head())
```

```python
df1.head()
```

```python
df1 = df1.drop (['experience_score_raw'], axis = 1)
```

```python
df1.head()
```

# model building

```python
x = df1.drop(['ExperienceScore'], axis = 1)
y = df1['ExperienceScore']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state =1)
```

```python
print("Shape of Training set : ", x_train.shape)
print("Shape of Training set : ", x_test.shape)
print(y_train.value_counts(normalize = True))
print("Percentage of classes in test set:")
print(y_test.value_counts(normalize = True))
```

```python
#training the model with logistic regression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
#print("\nROC-AUC Score:", roc_auc_score(y_test, model.predict_proba(x_test[:, 1])))
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

based on the confusion matrux, the true positives is 104, the true negative is 208, false positive is 44, false negative is 44

```python
#building model with Random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
#print("\nROC-AUC Score:", roc_auc_score(y_test, model.predict_proba(x_test[:, 1])))
accuracy = accuracy_score(y_test, y_pred_rf)
print("Model Accuracy:", accuracy)
```

```python
#buidling model with decison tree
```

```python
dt = DecisionTreeClassifier(random_state = 42)
dt.fit(x_train, y_train)
y_pred_dt = dt.predict(x_test)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
print("\nClassification Report:\n", classification_report(y_test, y_pred_dt))
#print("\nROC-AUC Score:", roc_auc_score(y_test, model.predict_proba(x_test[:, 1])))
accuracy = accuracy_score(y_test, y_pred_dt)
print("Model Accuracy:", accuracy)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

# Define the features you're using
features = [
    'Marital_Status',
    'Perception_Category',
    'Years_on_Treatment',
    'Overall_Comfort_Level',
    'Education_Level',
    'Monthly_Income',
    'Household_Size',
    'Treatment_Regimen',
    'Care_Timeliness',
    'Comfort_StaffInteraction'
]

# Create numeric input boxes for each feature
input_widgets = {}
for feature in features:
    input_widgets[feature] = widgets.FloatText(
        description=feature.replace("_", " ")[:18],  # Adjust label length for UI
        layout=widgets.Layout(width='400px')
    )

# Arrange the widgets in a vertical box
input_box = widgets.VBox([input_widgets[feature] for feature in features])

# Create the predict button and output area
predict_button = widgets.Button(description="Predict Experience Score", button_style='success')
output = widgets.Output()

# Define prediction function
def on_predict_clicked(b):
    with output:
        clear_output()
        # Gather inputs into DataFrame
        input_data = pd.DataFrame({feature: [input_widgets[feature].value] for feature in features})
        try:
            prediction = model.predict(input_data)[0]
            label_map = {1: "Positive Experience", 0: "Neutral Experience", 2: "Negative Experience"}
            print(f"Predicted Experience Score: {label_map.get(prediction, 'Unknown')} (Class: {prediction})")
        except Exception as e:
            print(f"Error during prediction: {e}")

# Link button click to function
predict_button.on_click(on_predict_clicked)

# Display all components
display(input_box, predict_button, output)
```
